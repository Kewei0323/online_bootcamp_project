## Abstract

This kernel builds on top of the comprehensive pre-processing performed in https://www.kaggle.com/juliencs/house-prices-advanced-regression-techniques/a-study-on-regression-applied-to-the-ames-dataset/

The main objectives of this Kernel include:

1) Categorical Variable Transformation using three separate methods:

a) Transform the categorical variables into a single variable using Ridge Regression. In this method, the training data is divided into "K" partitions. For each partition, prediction is computed by building model with other K-1 partitions. This aggregated prediction is then combined with numerical variables.

b) Encode categorical features using observation counts: The logic of the categorical count transformation lies in the fact that features with similar frequencies tend to behave similarly. 

c) Average All but One Encoding: In this method, a categorical feature value in a given sample is replaced by the mean of the responses for samples with same category excluding the sample under consideration. Logic: In case we find an extreme value or an outlier in a sample, the information provided by other similar data points with the same category in other samples could be still be used in that sample as well as other related information contained in the outlier's sample. 

I found the above three techniques shared by Owen Zhang, top 10 in Kaggle ranking.

2)  Then, I combine the categorical transformation with the numerical variables for cross-validation and hyper-parameter tuning of XGBoost, GBM

### Categorical Variable Transformation Strategy 1: Ridge Regression

In this section, I transform the categorical dummy variables to a single ridge Regression prediction variable.

It is important not to overfit to avoid this Ridge Regression variable from taking significant weight when later combined with the numerical variables.

```python
## Ridge Regression
from sklearn.linear_model import Ridge
from sklearn import cross_validation

state = 1
kf = KFold(n_splits=5, shuffle=False, random_state=state)
#train_cat_dummies['CatRidge'] = pd.Series(np.nan())

fold = 0
ridge_fold = Ridge()

## Methodology
# For training set, the model fit for the other K-1 partitions will be used for prediction
# For validation set, the mean of the predictions for the K models
CatRidge_train = np.zeros(y_train.shape[0])
CatRidge_valid = np.zeros((y_valid.shape[0],kf.get_n_splits()))

for train_index, test_index in skf.split(y_train):
    X_train_fold  = X_train_cat_dummy.iloc[train_index,:]
    X_test_fold   = X_train_cat_dummy.iloc[test_index,:]
    y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]

    ridge_fold.fit(X_train_fold, y_train_fold)
    CatRidge_train[test_index] = ridge_fold.predict(X_test_fold)
    CatRidge_valid[:,fold] = ridge_fold.predict(X_valid_cat_dummy)
    fold = fold + 1
    print("Fold = %d, Score = %f"% (fold, ridge_fold.score(X_test_fold,y_test_fold)))
```
